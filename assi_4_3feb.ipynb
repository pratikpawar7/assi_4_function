```python
'''Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
function.'''

gernater function is function that returns an iterator that produces a sequence of values when iterated over.
```

```python
yield keyword is used for produce a genrator value to the caller without terminator 
```

```python
# this ficonachi function is the best exaple of the genrator function 
def fibo(x):    
    a ,b = 0, 1
    for i in range(x):
        yield a 
        a , b = b , a+b 
        
for i in fibo(10):
    print(i)
```
